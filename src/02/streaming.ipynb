## 流式响应

流式传输意味着在生成模型时逐字显示模型的输出，而不是等待完整的响应。它让您的应用程序感觉更快、更活跃

**资源：**

1、OpenAI Streaming Docs（官方，免费）

链接：<https://platform.openai.com/docs/api-reference/streaming>

将stream=True添加到请求并迭代块的参考

2、 Anthropic Streaming Docs（官方，免费）

链接：<https://platform.claude.com/docs/zh-CN/build-with-claude/streaming>

Anthropic 的流 API 参考和 Python 示例

3、 Streaming LLM API 的工作原理 – Simon Willison（免费）

链接：<https://til.simonwillison.net/llms/streaming-llm-apis>

对 OpenAI、Anthropic 和 Google 的服务器发送事件如何在后台工作的清晰技术分解，有助于了解 HTTP 级别实际发生的情况

**关注点：** 设置stream=True、迭代增量块、组装各部分的完整响应以及使用 StreamingResponse 将流连接到 FastAPI 端点

**提示：** 对于面向用户的应用程序来说，流式传输几乎总是正确的选择。没有人愿意盯着加载旋转器 10 秒钟，等待立即出现完整响应

```typescript
import { OpenAI } from 'openai'

const client = new OpenAI()
```

```typescript
const stream = await client.responses.create({
  model: 'gpt-5.4-nano',
  input: [
    {
      role: 'user',
      content: '快速地说“双层泡泡浴”这个词十遍。',
    },
  ],
  stream: true,
})

for await (const event of stream) {
  console.log(event)
}
```
